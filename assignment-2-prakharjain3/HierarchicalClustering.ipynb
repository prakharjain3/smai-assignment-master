```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
import seaborn as sns  # Import Seaborn
from itertools import chain, combinations


# Subtask 1: 
class HierarchicalClustering:
    def __init__(self, data=None):
        self.data = data

    def linkages(self, linkage_type='single'):
        # Calculate linkage matrix
        linkage_matrix = linkage(self.data, method=linkage_type)
        return linkage_matrix

    def dendrogram(self, linkage_matrix):
        plt.figure(figsize=(12, 8))
        dendrogram(linkage_matrix, leaf_font_size=12, labels=self.data.index)  # Add labels and adjust font size
        plt.title('Hierarchical Clustering Dendrogram')
        plt.xlabel('Sample Index')
        plt.ylabel('Distance')
        plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
        sns.despine()  # Remove the top and right spines
        plt.show()

    def perform_clustering(self, linkage_type='single'):
        linkage_matrix = self.linkages(linkage_type)
        self.dendrogram(linkage_matrix)
        return linkage_matrix
```

```python
# Subtask 2:
# Perform hierarchical clustering on the dataset and obtain the linkage ma-
# trix. Vary the linkages and features used and state your observations.
# Plot the dendogram using the linkage matrix.

def powerset(iterable):
    xs = list(iterable)
    return chain.from_iterable(combinations(xs,n) for n in range(len(xs)+1))

data = pd.read_csv('SMAI-Dataset-hc-dataset/new_customers.csv')
data['Genre'] = data['Genre'].map(lambda x : 0 if x == 'Male' else 1)

X = data.columns.values

# create powerset of X


X_powerset = list(powerset(X))
X_powerset = X_powerset[1:]

# perform clustering on each powerset
k = 0
for i in range(len(X_powerset)):
    data_subset = data[list(X_powerset[i])]
    hc = HierarchicalClustering(data_subset)
    linkage_matrix = hc.perform_clustering('single')
    linkage_matrix = hc.perform_clustering('complete')
    linkage_matrix = hc.perform_clustering('average')
    linkage_matrix = hc.perform_clustering('weighted')
    linkage_matrix = hc.perform_clustering('centroid')
    linkage_matrix = hc.perform_clustering('median')
    linkage_matrix = hc.perform_clustering('ward')
    k += 1
    if k == 2:
        break
```
